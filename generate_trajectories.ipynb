```python
import numpy as np
from matplotlib import pyplot as plt
from trajectory_generator import TrajectoryGenerator

%load_ext autoreload
%autoreload 2
```

```python
T = 20  # duration of simulated trajectories (seconds)
srate = 50  # sampling rate (Hz)

border_region = 0.03  # max. distance to wall
sequence_length = T * srate  # number of steps in trajectory
box_width = 2.2       # width of training environment
box_height = 2.2      # height of training environment

trajectory_generator = TrajectoryGenerator(sequence_length, border_region)
positions, velocity, head_dir = trajectory_generator.generate_trajectory(box_width, box_height)
```

```python
plt.scatter(positions[0,0], positions[0,1], color = 'red', label = 'starting point')
plt.plot(positions[:,0], positions[:,1], label = 'simulated trajectory')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.legend(bbox_to_anchor=(1.45, 1.))
plt.show()

plt.savefig('sim_trajectories.eps', format='eps')
```

```python
n_samples = np.arange(0, sequence_length, 1) / srate
plt.plot(n_samples, velocity)
plt.title('Velocity')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.show()
```

```python
n_samples = np.arange(0, sequence_length, 1) / srate
plt.plot(n_samples, head_dir)
plt.title('Head direction')
plt.xlabel('Time (s)')
plt.ylabel('Head direction (rads)')
plt.show()
```
